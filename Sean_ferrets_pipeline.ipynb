```python
import pandas as pd
import numpy as np

ferrets = pd.read_csv("omkar/omkar.csv", header = 21)
ferrets
```

```python
ferrets.head()
```

```python
ferrets['subject.behavior.modifier'] = ferrets['Subject'] + ferrets["Behavior"] + ferrets["Modifier 1"]
```

```python
output = pd.DataFrame(columns = ["Subject", "Behavior", "Behavioral category", "Modifier", "Comment", "Start_Time", "End_Time", "Duration"])

for combination_key in pd.unique(ferrets['subject.behavior.modifier']):

    tmp = ferrets[ferrets['subject.behavior.modifier'] ==  combination_key]

    for i, g in tmp.groupby(np.arange(len(tmp)) // 2):
        #print(g["Subject"].iloc[0])
        tmp_dict = {
            'Subject': g["Subject"].iloc[0],
            'Behavior': g["Behavior"].iloc[0],
            'Behavioral category': g["Behavioral category"].iloc[0],
            'Modifier': g["Modifier 1"].iloc[0],
            'Comment': str(g["Comment"].iloc[0]) + str(g["Comment"].iloc[1]),
            'Start_Time': g["Time"].iloc[0],
            'End_Time': g["Time"].iloc[1],
            'Duration': g["Time"].iloc[1] - g["Time"].iloc[0]
        }   
        dct = {k:[v] for k,v in tmp_dict.items()} 
        output = pd.concat([output, pd.DataFrame(dct)], ignore_index = True)
        
output
```

```python
x = set(ferrets["Time"])
y = set(output["Start_Time"]).union(set(output["End_Time"]))
z = x.difference(y)
z
```
